# Class Ballancing 


```python
# importing Modules
from sklearn.utils import resample 
```

1- Up - Sampling minority Classes 

```python
mData.info()
```

```python
Categories = pd.Categorical(mData['quality']).categories
numOfCategories = pd.Categorical(mData['quality']).categories.size
CategorySizes = mData.quality.value_counts()
maxSize = mData.quality.value_counts().max()
mSampledDate =[]
for Category in Categories :
    if (mData[mData.quality==Category].shape[0] != maxSize):
        temp = resample(mData[mData.quality==Category], 
                                 replace=True,     # sample with replacement
                                 n_samples=(maxSize-mData[mData.quality==Category].shape[0]),    
                                 random_state=123) # reproducible results
        mSampledDate.append(temp)

BallancedData=mData
for data in mSampledDate:
    BallancedData = pd.concat([BallancedData, data])
BallancedData.quality.value_counts()
BallancedData
```

# Clustering Using K mean (Original Data)
